```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
def standard_csv(dataset):  # 除了2020-2022 human
    number = {}
    array = np.arange(1,10000)
    array_ = [str(x) + '. ' for x in array]
    array_dict = dict(zip(array_,array_))
    human_info = open(dataset,encoding='utf_8_sig')
    human_info = human_info.read()
    human_info_list = human_info.split('\n\n')
    human_info_list = [x.replace('\n',' ').strip() for x in human_info_list]
    number = []
    for value in human_info_list:
        if str(value[:2]) in array_dict:
            number.append(value)
        elif str(value[:3]) in array_dict:
            number.append(value)
        elif str(value[:4]) in array_dict:
            number.append(value)
        elif str(value[:5]) in array_dict:
            number.append(value)
        elif str(value[:6]) in array_dict:
            number.append(value)
        else:
            pass

    data_human = pd.DataFrame({'info':number})
    data1_human = pd.DataFrame({'info':human_info_list})
    index_list = []
    for key in data_human['info'].to_list():
        index_list.append(data1_human[data1_human['info'] == key].index[0])
    context_list = []
    for i in range(len(index_list)-1):
        df_mark = data1_human.iloc[index_list[i]:index_list[i+1]]
        context_list.append('\t'.join(df_mark['info'].tolist()))
    
    df_mark = data1_human.iloc[index_list[-1]:]
    context_list.append('\t'.join(df_mark['info'].tolist()))
    df_final = pd.DataFrame({'info':context_list})
    return df_final

target_dir_list = ['./lung_bstract-singlecell-set_title.txt']
#target_dir_list = ['./abstract-singlecell-set.txt']
tmp = 0
for target_dir in  target_dir_list:
    abstract_singlecell_set = standard_csv(target_dir)
    if type(tmp) == int:
        tmp = abstract_singlecell_set
    else:
        tmp = pd.concat([tmp,abstract_singlecell_set])
    tmp.drop_duplicates(inplace=True)
tmp
```

```python
# abstract的匹配方法
def get_abstract_info(x):
    a_ = x.split('\t')
    for a in a_:
        if 'pmid' in a or 'PMID' in a:
            result = a_[a_.index(a)-1]
            if '©' not in result:
                if "Author information" not in result and "Comment on" not in result and "Comment in" not in result:
                    return result
                else:
                    return '-'
            else:
                if "Author information" not in a_[a_.index(a)-2] and "Comment on" not in a_[a_.index(a)-2] and "Comment in" not in result:
                    return a_[a_.index(a)-2]
                else:
                    return '-'


def standard_input(result):
    result['info'] = result['info'].replace({'pmid':'PMID','doi':'DOI','pmcid':'PMCID'})
    result['pmid'] = result['info'].map(lambda x: re.findall('PMID:(.*)',x)[0].split(' ')[1].strip().split('\t')[0].strip())
    result['doi'] = result['info'].map(lambda x: 'DOI' + x.split('DOI')[-1].split('PMCID')[0].split('PMID')[0].strip())
    result['pmcid'] = result['info'].map(lambda x:"PMCID" + x.split('PMCID')[-1].split('DOI')[0].split('PMID')[0].strip())
    result.loc[[value for value in result['doi']  if value[:3] != 'DOI'],'doi'] = '-'
    result.loc[[value for value in result['pmcid']  if value[:5] != 'PMCID'],'pmcid'] = '-'
    result['pmid'] = result['pmid'].map(lambda x:x.split('.')[0].split(';')[0].strip())
    result['title'] = result['info'].map(lambda x:x.split('\t')[1].strip())
    result['data'] = result['info'].map(lambda x:x.split('.')[2].split(':')[0].split(';')[0].strip())
    result['journal_name'] = result['info'].map(lambda x:x.split('.')[1].strip())
    result['pmcid'] = result['pmcid'].map(lambda x:x if len(x) < 20 else '-')
    result['doi'] = result['doi'].map(lambda x:x if len(x) < 40 else '-')
    result['abstract'] =  result['info'].map(get_abstract_info)
    return result


abstract_singlecell_set_2016_2022 = standard_input(tmp)
abstract_singlecell_set_2016_2022.drop(columns=['info'],inplace=True)
# abstract_singlecell_set_2016_2022.to_excel('./abstract_singlecell_set_2016_2022.xlsx',index=False,encoding='utf_8_sig')
abstract_singlecell_set_2016_2022.drop_duplicates(subset='pmid',inplace=True)
abstract_singlecell_set_2016_2022
```
