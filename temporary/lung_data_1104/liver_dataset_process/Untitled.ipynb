```python
import pandas as pd
```

```python
df = pd.read_csv('./info_liuke_1122.csv')
df.head()
```

```python
df_ = pd.read_excel('./01.pubmed模板_liver_1116.xlsx',sheet_name='数据')
df_['pmid'] = df_['pmid'].astype(str)
df['pmid'] = df['pmid'].astype(str)
```

```python
df_.columns


article_dict = dict(zip(df_['pmid'],df_['article']))
journal_dict = dict(zip(df_['pmid'],df_['journal']))
DOI_dict = dict(zip(df_['pmid'],df_['DOI']))
abstract_dict = dict(zip(df_['pmid'],df_['abstract']))
article_pubdate_dict = dict(zip(df_['pmid'],df_['article_pubdate']))


df['article'] = df['pmid'].map(article_dict)
df['journal'] = df['pmid'].map(journal_dict)
df['DOI'] = df['pmid'].map(DOI_dict)
df['abstract'] = df['pmid'].map(abstract_dict)
df['article_pubdate'] = df['pmid'].map(article_pubdate_dict)
```

```python
df
```

```python
df.to_csv('./fsafsfsfs.csv',index=False,encoding='utf_8_sig')
```
