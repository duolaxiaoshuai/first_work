### 标准化&数据提取

```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
def standard_csv(dataset):  # 除了2020-2022 human
    number = {}
    array = np.arange(1,10000)
    array_ = [str(x) + '. ' for x in array]
    array_dict = dict(zip(array_,array_))
    human_info = open(dataset,encoding='utf_8_sig')
    human_info = human_info.read()
    human_info_list = human_info.split('\n\n')
    human_info_list = [x.replace('\n',' ').strip() for x in human_info_list]
    number = []
    for value in human_info_list:
        if str(value[:2]) in array_dict:
            number.append(value)
        elif str(value[:3]) in array_dict:
            number.append(value)
        elif str(value[:4]) in array_dict:
            number.append(value)
        elif str(value[:5]) in array_dict:
            number.append(value)
        elif str(value[:6]) in array_dict:
            number.append(value)
        else:
            pass

    data_human = pd.DataFrame({'info':number})
    data1_human = pd.DataFrame({'info':human_info_list})
    index_list = []
    for key in data_human['info'].to_list():
        index_list.append(data1_human[data1_human['info'] == key].index[0])
    context_list = []
    for i in range(len(index_list)-1):
        df_mark = data1_human.iloc[index_list[i]:index_list[i+1]]
        context_list.append('\t'.join(df_mark['info'].tolist()))
    
    df_mark = data1_human.iloc[index_list[-1]:]
    context_list.append('\t'.join(df_mark['info'].tolist()))
    df_final = pd.DataFrame({'info':context_list})
    return df_final

#target_dir_list = ['./abstract-singlecell-set_2016_2019.txt','./abstract-singlecell-set_2019_2021.txt','./abstract-singlecell-set_2021_2022.txt']
target_dir_list = ['./liver_abstract-singlecell-set.txt']
tmp = 0
for target_dir in  target_dir_list:
    abstract_singlecell_set = standard_csv(target_dir)
    if type(tmp) == int:
        tmp = abstract_singlecell_set
    else:
        tmp = pd.concat([tmp,abstract_singlecell_set])
    tmp.drop_duplicates(inplace=True)
tmp
```

```python
# 不一定需要这一步，在总数量对应不上的时候，可以做这一步操作
def check_number_fun(dataset):
    flag, info_list  = 1 , []
    for value in tmp['info'].values:
        if int(value.split('.')[0].strip()) == flag:
            info_list.append(value)
            flag += 1
        else:
            info_list[-1] = info_list[-1] + '\t' + value
    tmp = pd.DataFrame({'info':info_list})
tmp
```

```python
# abstract的匹配方法
def get_abstract_info(x):
    a_ = x.split('\t')
    for a in a_:
        if 'pmid' in a or 'PMID' in a:
            result = a_[a_.index(a)-1]
            if '©' not in result:
                if "Author information" not in result and "Comment on" not in result and "Comment in" not in result:
                    return result
                else:
                    return '-'
            else:
                if "Author information" not in a_[a_.index(a)-2] and "Comment on" not in a_[a_.index(a)-2] and "Comment in" not in result:
                    return a_[a_.index(a)-2]
                else:
                    return '-'


def standard_input(result):
    result['info'] = result['info'].replace({'pmid':'PMID','doi':'DOI','pmcid':'PMCID'})
    result['pmid'] = result['info'].map(lambda x: re.findall('PMID:(.*)',x)[0].split(' ')[1].strip().split('\t')[0].strip())
    result['doi'] = result['info'].map(lambda x: 'DOI' + x.split('DOI')[-1].split('PMCID')[0].split('PMID')[0].strip())
    result['pmcid'] = result['info'].map(lambda x:"PMCID" + x.split('PMCID')[-1].split('DOI')[0].split('PMID')[0].strip())
    result.loc[[value for value in result['doi']  if value[:3] != 'DOI'],'doi'] = '-'
    result.loc[[value for value in result['pmcid']  if value[:5] != 'PMCID'],'pmcid'] = '-'
    result['pmid'] = result['pmid'].map(lambda x:x.split('.')[0].split(';')[0].strip())
    result['title'] = result['info'].map(lambda x:x.split('\t')[1].strip())
    result['data'] = result['info'].map(lambda x:x.split('.')[2].split(':')[0].split(';')[0].strip())
    result['journal_name'] = result['info'].map(lambda x:x.split('.')[1].strip())
    result['pmcid'] = result['pmcid'].map(lambda x:x if len(x) < 20 else '-')
    result['doi'] = result['doi'].map(lambda x:x if len(x) < 40 else '-')
    result['abstract'] =  result['info'].map(get_abstract_info)
    return result


abstract_singlecell_set_2016_2022 = standard_input(tmp)
abstract_singlecell_set_2016_2022.drop(columns=['info'],inplace=True)
# abstract_singlecell_set_2016_2022.to_excel('./abstract_singlecell_set_2016_2022.xlsx',index=False,encoding='utf_8_sig')
abstract_singlecell_set_2016_2022
```

### 筛选过程  ['single cell','atlas','single-cell']   和 ['lung','respiratory','pneumonia','pulmonary','pneuma','pulmonic']

```python
# ((lung) OR (respiratory) OR (pneumonia) OR (pulmonary) OR (pneuma) OR (pulmonic))
# a = ['lung','respiratory','pneumonia','pulmonary','pneuma','pulmonic']
#((single cell) OR (atlas)) AND((mouse) OR (human)) AND ((lung) OR (respiratory) OR (pneumonia) OR (pulmonary) OR (pneuma) OR (pulmonic))

def fun_filter_1(x):
    result = []
    x = x.split(' ')
    for value in x:
        result.append(value.lower().strip())
    mark1 = ['single cell','atlas','single-cell']
    count = 0
    for value_ in result:
        if mark1[0] in value_ or mark1[1] in value_ or mark1[2] in value_:
            count += 1
    if count >= 1:
        return 1
    return 0

def fun_filter_2(x):  #  舍弃
    result = []
    x = x.split(' ')
    for value in x:
        result.append(value.lower().strip())
    mark2 = ['mouse','human']
    count = 0
    for value_ in result:
        if mark2[0] in value_ or mark2[1] in value_:
            count += 1
    if count >= 1:
        return 1
    return 0

        
def fun_filter_3(x):
    result = []
    x = x.split(' ')
    for value in x:
        result.append(value.lower().strip())
#     mark3 = ['lung','respiratory','pneumonia','pulmonary','pneuma','pulmonic']
    mark3 = ['liver','hepar','hepatic','hepatitis']
    count = 0
    for value_ in result:
        if mark3[0] in value_ or mark3[1] in value_ or mark3[2] in value_ or mark3[3] in value_:
            count += 1
    if count >= 1:
        return 1
    return 0
abstract_singlecell_set_2016_20221['mark_'] = abstract_singlecell_set_2016_20221['title'] + ' ' + abstract_singlecell_set_2016_20221['abstract']
abstract_singlecell_set_2016_20221['mark1'] = abstract_singlecell_set_2016_20221['mark_'].map(fun_filter_1)
abstract_singlecell_set_2016_20221['mark3'] = abstract_singlecell_set_2016_20221['mark_'].map(fun_filter_3)
abstract_singlecell_set_2016_20221['mark_final'] = abstract_singlecell_set_2016_20221['mark1'] + abstract_singlecell_set_2016_20221['mark3']
#abstract_singlecell_set_2016_20221[abstract_singlecell_set_2016_20221['mark'] == 1].to_excel("./liver_result_1107.xlsx",index=False,encoding='utf_8_sig')
abstract_singlecell_set_2016_20221[abstract_singlecell_set_2016_20221['mark_final'] == 2]
```

### mesh_term_keywords的处理过程

```python
class mash_term_cal:
    def fun(x):
        a_list = x.split(';')
        a = []
        for key in a_list:
            if len(key.split('/')) == 1:
                a.append(key.strip())
        a_ = set(a)
        return ','.join(list(a_))

    def fun1(x):
        a_list = x.split(';')
        b = []
        for key in a_list:
            if len(key.split('/')) == 1:
                pass
            else:
                b.append(key.split('/')[0].strip())
        b_ = set(b)
        return ','.join(list(b_))

    def fun2(x):
        if x[0] == ',':
            return x[1:]
        elif x[-1]:
            return x[:-1]
        return x

    def mesh_key_values(target_file):
        df = target_file
        df1 = df[['pmid', 'mesh_term']]
        df1['mesh_term'].fillna('-', inplace=True)
        df1['x'] = df1['mesh_term'].map(self.fun)
        df1['y'] = df1['mesh_term'].map(self.fun1)

        df_mesh_x = pd.read_excel('./mesh_term_deal/mesh_standard.xlsx', sheet_name='x')
        df_mesh_y = pd.read_excel('./mesh_term_deal/mesh_standard.xlsx', sheet_name='y')
        result_x, result_y = [], []
        x_list = df_mesh_x['x'].tolist()
        for key in df1['x'].tolist():
            key_list = key.split(',')
            a_x = []
            for i in key_list:
                if i.strip() not in x_list:
                    a_x.append(i.strip().split('*')[0])
            result = ','.join(a_x)
            result_x.append(result)
        y_list = df_mesh_y['y'].tolist()
        for key in df1['y'].tolist():
            key_list = key.split(',')
            a_y = []
            for i in key_list:
                if i.strip() not in y_list:
                    a_y.append(i.strip().split('*')[0])
            result = ','.join(a_y)
            result_y.append(result)
        df1['result_x'] = result_x
        df1['result_y'] = result_y
        df1['result_final'] = df1['result_x'] + ',' + df1['result_y']
        df1 = df1[['pmid', 'mesh_term', 'result_final']]
        df1['result_final'] = df1['result_final'].map(self.fun2)
        df1.replace({'-': np.NAN}, inplace=True)
        mesh_key_values_dict = dict(zip(df1['pmid'], df1['result_final']))
        target_file['mesh_key_values'] = target_file['pmid'].map(mesh_key_values_dict)
        return target_file
```

### 爬虫之后的文件结果，补充的journal_name_full_name, author, associated_data, keywords, mersh_term信息

```python
# 数据清洗
def data_cleaning(dataset):
    df_ = dataset
    df_['doi'] = df_['doi'].map(lambda x:x.replace(': ',':'))
    df_['pmcid'] = df_['pmcid'].map(lambda x:'-' if x[:6] != 'PMCID:' else x)
    df_['doi'] = df_['doi'].map(lambda x:'-' if x[:4] != 'DOI:' else x)
    df_['author'] = df_['author'].map(lambda x:x.split("']")[0].strip())
    df_['journal_name_full_name'] = df_['journal_name_full_name'].map(lambda x:x.replace('amp; ',' '))
    df_['journal_name_full_name'] = df_['journal_name_full_name'].map(lambda x:x.split(':')[0].strip())
    df_['journal_name_full_name'] = df_['journal_name_full_name'].map(lambda x:x.split('=')[0].strip())
    df_['journal_name_full_name'] = df_['journal_name_full_name'].map(lambda x:x.replace('#x27;s',' '))
    df_['journal_name_full_name'] = df_['journal_name_full_name'].map(lambda x:x + ')' if '(' in x and ')' not in x else x)
    df_['journal_name_full_name'] = df_['journal_name_full_name'].map(lambda x:x + ')' if '(' in x and ')' not in x else x)
    df_['journal_name_full_name'] = df_['journal_name_full_name'].map(lambda x:x.split('(')[0].strip())

    df_result1  = df_
    df_result1['journal_name_'] = df_result1['journal_name_full_name'].map(lambda x:x.split('(')[0].split(':')[0].lower().strip())
    df_result1['journal_name_'] = df_result1['journal_name_'].map(lambda x:x.split('[')[0].split(':')[0].lower().strip())

    df_2022 = pd.read_excel('./2022(1).xlsx')
    df_2022['journal_name_'] = df_2022['journal_name'].map(lambda x:x.lower().strip())

    if_2022_dict = dict(zip(df_2022['journal_name_'],df_2022['if_2022']))
    df_result1['if_2022'] = df_result1['journal_name_'].map(if_2022_dict)
    df_result1.drop(columns=['journal_name_'],inplace=True)
    return df_result1



# filter 数据 microRNA，LongmicroRNA，ATAC，circRNA，BulkRNA 影响因子 >= 10
def filter(dataset):
    dataset = dataset[dataset['if_2022'] >= 10]
    return dataset

def to_supplement_info(source_file,mark_file_dir):
    df1 = source_file
    df1['pmid'] = df1['pmid'].astype('int')
    
    df_ = pd.read_csv(mark_file_dir,header=None)
    df_.columns = ['pmid','journal_name_full_name', 'author','associated_data','keywords','mersh_term']
    df_['pmid'] = df_['pmid'].astype('int')
    journal_name_full_name_dict = dict(zip(df_['pmid'],df_['journal_name_full_name']))
    
    author_name_dict = dict(zip(df_['pmid'],df_['author']))
    keywords_data_dict = dict(zip(df_['pmid'],df_['keywords']))
    mersh_term_dict = dict(zip(df_['pmid'],df_['mersh_term']))
    
    df1['journal_name_full_name'] = df1['pmid'].map(journal_name_full_name_dict)
    df1['author'] = df1['pmid'].map(author_name_dict)
    df1['keywords'] = df1['pmid'].map(keywords_data_dict)
    df1['mesh_term'] = df1['pmid'].map(mersh_term_dict)
    
    # 数据清洗
    df1 = data_cleaning(df1)
    df1 = df1[['journal_name', 'journal_name_full_name','if_2022','pmid', 'doi', 'pmcid', 'data','title', 'abstract', 'author', 'keywords', 'mesh_term']]
  
#     df1 = mesh_key_values(df1)

    # 筛选影响因子大于等于10的文章
#     df1 = filter(df1)
    return df1

df_result = to_supplement_info(abstract_singlecell_set_2016_2022,'./resut_1104.csv')

# df_result.to_excel('./lung_result_1104gdsgsd.xlsx',index=False,encoding='utf_8_sig')
df_result.head()
```
